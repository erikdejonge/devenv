```python
import requests
c = requests.get(url='https://feeds.megaphone.fm/LM1344278906').text
```

```python
import feedparser
f = feedparser.parse(c)
```

```python
s = set()
for i in f["entries"]:
    for l in i["links"]:
        s.add(l.href)
        
cnt = 0
for i in s:
    print("wget "+i+" &")
    cnt += 1
    if cnt % 10==0:
        print("wait")
        cnt = 0
```
